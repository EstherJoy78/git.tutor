```python
#Aggregating Data
#Using groupby for data aggregation.
df = pd.read_csv('retail_store_1jtq.csv')
df
df.head()

```

```python
#New column
df['Sales'] = df['Price'] * df['QuantitySold']
df.head()
```

```python
#groupby 'category' and calculate total sales.
grouped_df = df.groupby('Category')['Sales'].sum().reset_index()
grouped_df
```

```python
#Using a Pivot table for data aggregation.
df = pd.read_csv('retail_store_1jtq.csv')
df.head()
```

```python
pivot_table = pd.pivot_table(df, values=['Price','QuantitySold'], index= 'Category', aggfunc= 'sum')
pivot_table
```

```python
#Visualizing Data.
#import matplotlib.pyplot as plt to perform data visualization.
import matplotlib.pyplot as plt
```

```python
#bar charts.
df = pd.read_csv('retail_store_1jtq.csv')
df.head()
#count the number of occurence for each category
category_counts = df['Category'].value_counts()
category_counts
```

```python
#bar chart creation
plt.bar(category_counts.index,category_counts.values)
plt.title('Quantity of each category in Retail Store')
plt.xlabel('Category')
plt.ylabel('Quantity')

plt.xticks(rotation= 45)
plt.show()
```

```python
#histogram
#It is an estimate of a probability distribution of a continuous variable.
#the data is divided into a set of bins, and the count of observations that fall into each bin is represented by the height of a bar. Histograms are commonly used to visualize the frequency or density of data in different intervals orc ranges.
df = pd.read_csv('retail_store_1jtq.csv')
df.head()
```

```python
plt.hist(df['Product_Sales Histogram'], bins= 10)
plt.title('Product_Sales Histogram')
plt.xlabel('Product')
plt.ylabel('QuantitySold')
plt.show()
```
