# AVIATION_ACCIDENTS ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AviationData.csv', encoding = 'cp1252')
df.head()
```

```python
assert type(df) == pd.DataFrame
```

```python
df.columns
```

```python
relevant_columns = [
    "Event.Id",
    "Investigation.Type",
    "Event.Date",
    "Location",
    "Country",
    "Injury.Severity",
    "Aircraft.damage",
    "Make",
    "Model",
    "Number.of.Engines",
    "Purpose.of.flight",
    "Total.Fatal.Injuries",
    "Total.Serious.Injuries",
    "Total.Minor.Injuries",
    "Total.Uninjured",
    "Weather.Condition"
]
```

```python
df = pd.DataFrame(df,columns = relevant_columns)
df.head()
```

```python
df['Fatality'] = df['Injury.Severity'].str.extract(r'\((\d+)\)')
df['Fatality'].replace({'Non-Fatal': 0, 'Minor': 0, 'Serious':0, 'Incident':0},inplace=True)
df['Fatality'].fillna(df['Injury.Severity'] , inplace=True)
df.replace('Unavailable',np.nan, inplace=True)
df['Fatality'] = pd.to_numeric(df['Fatality'], errors = 'coerce').fillna(0).astype(int)
df['Fatality'] = df.apply(lambda row: row['Total.Fatal.Injuries'] if 'Fatal' in str(row['Injury.Severity'])else row ['Fatality'],axis=1)
pd.options.display.float_format = '{:.0f}'.format
df.head()
```

```python
df['Fatality'] = pd.to_numeric(df['Fatality'], errors = 'coerce').astype(object)
df.head()
```

```python
df['Fatality'].value_counts()
```

```python
df.drop(columns = ['Total.Fatal.Injuries'], inplace= True)
```

```python
assert df['Fatality'].dtype == object
```

```python
#Country
df['Country'].value_counts()
```

```python
df_ust = df[df['Country'] == 'United States']
df_br = df[df['Country'] == 'Brazil']
df_ust, df_br.reset_index(drop=True, inplace= True)
df.head()
```

```python
df_State_Codes = 'State_Code'
df_State_Codes = pd.read_csv('USState_Codes.csv')
```

```python
State_Code = [
    'AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA',
    'HI','ID','IL','IN','IA','KS','KY','LA','ME','MD',
    'MA','MI','MN','MS','MO','MT','NE','NE','NH','NJ',
    'NM','NY','NC','ND','OH','OK','OR','PA','RI','SC',
    'SD','TN','TX','UT','VT','VA','WA','WV','WI','WY'
]
df_ust['Event.Date'] = pd.to_datetime(df_ust['Event.Date'], format='%Y-%m-%d')
df_ust['Month'] = df_ust['Event.Date'].dt.month
df_ust['Year'] = df_ust['Event.Date'].dt.year
def extract_City_State_Code(Location):
    if pd.notna(Location):
       Location = Location.strip()
       last_two_chars = Location[-2:].upper()
       if last_two_chars in State_Code:
        return Location[:-3].strip(),last_two_chars
       
    else:
        return Location, 'Not Applicable'
    
df_ust[['City','State']] = df_ust['Location' ].apply( extract_City_State_Code).apply(pd.Series)
df.head()
```

```python
df['Event.Date'] = pd.to_datetime(df['Event.Date'], format='%Y-%m-%d')
df['Month'] = df['Event.Date'].dt.month
df.head()
```

```python
df['Year'] = df['Event.Date'].dt.year
df.head()
```

```python
years = df_ust[df_ust['Year']>= 1982]
Top_10_states = years['State'].value_counts().head(10)
colors = plt.cm.tab10(range(len(Top_10_states)))

plt.figure(figsize=(6,4))
bars = plt.bar(Top_10_states.index, Top_10_states.values, color= colors)
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Top 10 states with most Accidents from 1982 onwards')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle= '--')

plt.show
```

```python
df['Year'] = df['Event.Date'].dt.year
df_ust = df_ust[df_ust['Year']>= 1982]
accidents_by_year = df_ust['Year'].value_counts().sort_index()

plt.figure(figsize=(8,4))
plt.plot(accidents_by_year.index, accidents_by_year.values, linestyle='-')
plt.title('Number of Aviation Accidents from 1982 onwards')
plt.xlabel('Year')
plt.ylabel('No of Accidents')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
df['Month'] = df['Event.Date'].dt.month
Accidents = df[df['Investigation.Type'] == 'Accident']

Month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

Number_of_Accidents = Accidents.groupby('Month')['Investigation.Type'].count()

Number_of_Accidents = Number_of_Accidents.reindex(range(1, 13), fill_value=0)

fig = plt.figure(figsize=(10, 6))
axes = plt.subplot(1, 1, 1)

axes.set_title('Distribution of Accidents by Month')
axes.set_xlabel('Month')
axes.set_ylabel('Number of Accidents')

axes.set_xticks(range(12))
axes.set_xticklabels(Month)

axes.plot(range(12), Number_of_Accidents)

plt.show()
```

```python
df['Year'] = df['Event.Date'].dt.year
df_ust= df[df['Year']>= 1982]
Yearly_stats = df_ust.groupby('Year')['Fatality'].agg(['count','sum'])
Yearly_stats['Fatality_Percentage'] = (Yearly_stats['sum']/Yearly_stats['count']) * 10
Yearly_stats['Fatality_Percentage'].fillna(-np.inf, inplace=True)

Year_with_the_highest_percentage = Yearly_stats['Fatality_Percentage'].idxmax()
highest_percentage = Yearly_stats['Fatality_Percentage'].max()

print(f'Year with the highest percentage of fatalities(1982 onwards):{Year_with_the_highest_percentage}')
print(f'highest percentage of fatalities:{highest_percentage:.2f}%')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(Yearly_stats.index, Yearly_stats['Fatality_Percentage'], marker='o', color='b', linestyle='-')
plt.annotate(f'{highest_percentage:2f}',
            xy=(Year_with_the_highest_percentage,highest_percentage),
            xytext=(Year_with_the_highest_percentage,highest_percentage+5),
            arrowprops=dict(facecolor='black', shrink=0.05), 
            ha='center')

plt.title('Yearly Percentage of Fatalities(1982 onwards)')
plt.xlabel('Year')
plt.ylabel('Fatality Percentage')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
df_ust['Weather.Condition'] = df_ust['Weather.Condition'].str.upper()
weather_counts= df_ust['Weather.Condition'].value_counts()

plt.figure(figsize=(10,6))
weather_counts.plot(kind= 'bar', color = 'blue')

plt.title('Distribution of Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Count')

plt.xticks(rotation= 45)
plt.grid(axis = 'y')
plt.tight_layout()
plt.show()
```

```python
make_model_counts = df_ust.groupby(['Make', 'Model']).size().reset_index(name= 'AccidentCount')
sorted_make_model_counts = make_model_counts.sort_values(by= 'AccidentCount', ascending= False)
sorted_make_model_counts
```

```python
df_ust['Number.of.Engines'] = df_ust['Number.of.Engines'].replace(0.0, np.nan)

# Group by 'Make' and 'Model' and count occurrences
make_model_accident_counts = df_ust.groupby(['Make', 'Model']).size().reset_index(name='AccidentCount')
make_model_accident_counts = make_model_accident_counts.sort_values(by='AccidentCount', ascending=False)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Top 10 Makes and Models by Accident Count
top_10_make_model = make_model_accident_counts.head(10)
axes[0].barh(top_10_make_model['Make'] + ' ' + top_10_make_model['Model'], top_10_make_model['AccidentCount'], color='skyblue')
axes[0].set_xlabel('Number of Accidents')
axes[0].set_ylabel('Make and Model')
axes[0].set_title('Top 10 Aircraft Makes and Models Involved in Accidents')
axes[0].invert_yaxis()

# Engine Types Distribution
engines_counts = df_ust['Number.of.Engines'].dropna().value_counts()
axes[1].bar(engines_counts.index, engines_counts.values, color='lightcoral')
axes[1].set_title('Distribution of Aircraft Engine Types')
axes[1].set_xlabel('Engine Type')
axes[1].set_ylabel('Number.of.Accidents')
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

# Conclusions.

From the observations, it can be concluded that favorable weather conditions do not necessarily prevent accidents in aviation. This suggests that other factors, such as pilot error, mechanical failure, or operational issues, play a significant role in aviation accidents.

Although there has been a reduction in the overall number of aviation accidents, California, Texas, and Florida are states where accidents occur more frequently. 

Additionally, the fact that specific aircraft makes, models, and engine types are involved in accidents more frequently suggests potential design or operational vulnerabilities with these aircraft. This could highlight the need for more stringent maintenance, inspections, or operational protocols for those particular types of aircraft.
